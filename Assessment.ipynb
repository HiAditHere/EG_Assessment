I have used the following approach for data engineering

- Seperate the dataframe into 2 parts, clean data frame with no '?' and unclean dataframe with '?' values.

- Convert the object type columns in clean dataframe to int type columns. After this, group the dataframe by LeagueIndex and find the mean values of all variables for every LeagueIndex.

- Use the mean values to replace the '?' in the unclean dataframe.

- As there are no age, hours per week and total hours for players in league index 8, I have made an assumption that all these players will have similar values to players in league index 7. Players at the top would have similar statistics to that of players just above or below them. So for any player with league index 8, I have replaced the '?' values with mean of players with league index 7.

```python
columns = ['Age', 'HoursPerWeek', 'TotalHours']

clean_df = df.drop(l)
clean_df['Age'] = clean_df['Age'].astype(str).astype(int)
clean_df['HoursPerWeek'] = clean_df['HoursPerWeek'].astype(str).astype(int)
clean_df['TotalHours'] = clean_df['TotalHours'].astype(str).astype(int)

mean = clean_df.groupby('LeagueIndex').mean()
```

```python
unclean_df = df.loc[l]
```

```python
mean['Age']
```

```python
columns = ['Age', 'HoursPerWeek', 'TotalHours']

for i in columns:
    for j in l:
        if(unclean_df[i][j] == '?'):
            if(unclean_df['LeagueIndex'][j] == 8):    
                unclean_df[i][j] = mean[i][7]
            else:
                unclean_df[i][j] = mean[i][df['LeagueIndex'][j]]
```

The object type columns in the unclean dataframe can lead to a problem later on while normalizing the data. The values can go to i'nf' if we merge the 2 dataframes and find mean and standard deviation as it is. So lets convert the object columns to float

```python
unclean_df['Age'] = unclean_df['Age'].astype(float)
unclean_df['TotalHours'] = unclean_df['TotalHours'].astype(float)
unclean_df['HoursPerWeek'] = unclean_df['HoursPerWeek'].astype(float)
```

Now lets test all the classification algorithms and check which one give the best results

1) Random Forest <br>
2) Logistic Regression <br>
3) SVM <br>
4) Gaussian Naive Bayes <br>
5) Gradient Boosting <br>
6) Decision Tree <br>
7) AdaBoost Classifier <br>

I have implemented a for loop. At every loop the model will be splitted into a new train and test set and fit using the above mentioned classifiers. The scores of all models at every loop are stored in a dictionary. At the end, an avg for all classifiers is returned. The classifier with the highest avg can be chosen.

Although we can use standard scaler for standardization, it is usually a good practice to use the mean and standard deviation of the train set to standardize the test set. If you were to calculate the mean and standard deviation of the test set and use them for standardization, you would be incorporating information from the test set into the training process. This violates the principle of maintaining a clear separation between the training and test data. This is called Data Leakage

```python
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

scores = {
    'Random Forest':[],
    'Logistic Regression':[],
    'SVM':[],
    'Gaussian Naive Bayes':[],
    'Gradient Boosting':[],
    'Descision Tree':[],
    'Ada Boost':[],
}

for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    mean = np.mean(X_train, axis = 0)
    std = np.std(X_train, axis = 0)

    X_train = (X_train-mean)/std
    X_test = (X_test-mean)/std
    
    clf1 = RandomForestClassifier(random_state = 0)
    clf1.fit(X_train, y_train)
    scores['Random Forest'].append(clf1.score(X_test, y_test))
    
    clf2 = LogisticRegression(random_state = True)
    clf2.fit(X_train, y_train)
    scores['Logistic Regression'].append(clf2.score(X_test, y_test))

    clf3 = SVC(gamma = 'auto')
    clf3.fit(X_train, y_train)
    scores['SVM'].append(clf3.score(X_test, y_test))

    clf4 = GaussianNB()
    clf4.fit(X_train, y_train)
    scores['Gaussian Naive Bayes'].append(clf4.score(X_test, y_test))

    clf5 = GradientBoostingClassifier()
    clf5.fit(X_train, y_train)
    scores['Gradient Boosting'].append(clf5.score(X_test, y_test))

    clf6 = DecisionTreeClassifier(random_state = 0)
    clf6.fit(X_train, y_train)
    scores['Descision Tree'].append(clf6.score(X_test, y_test))

    clf7 = AdaBoostClassifier()
    clf7.fit(X_train, y_train)
    scores['Ada Boost'].append(clf7.score(X_test, y_test))
    
for i in scores:
    print(i, end = ":")
    for j in scores[i]:
        print(round(j,2), end = " ")
    print()
    print('Avg = ' + str(sum(scores[i])/5))
    print()
```

Thus we can conclude that Random Forest would be the best classifier in this case
